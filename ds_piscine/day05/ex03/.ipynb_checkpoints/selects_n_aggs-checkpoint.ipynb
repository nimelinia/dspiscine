```python
import pandas as pd
```

<h2><center>Создание dataframe</center></h2>

```python
df = pd.read_json('../data/auto.json', orient='records')
```

```python
df.head()
```

```python
df.set_index('CarNumber', inplace=True)
```

```python
df.head()
```

<h2><center>Отбор по параметрам</center></h2>

```python
df[df['Fines'] > 2100]
```

```python
df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

```python
df[df['Model'].isin(['Focus', 'Corolla'])]
```

```python
df[df.index.isin(['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS'])]
```

<h2><center>Агрегация с Make & Model</center></h2>

```python
df.groupby('Make').median().Fines
```

```python
df.groupby(['Make', 'Model']).median().Fines
```

```python
df.groupby(['Make', 'Model']).count().Fines
```

```python
df.groupby(['Make', 'Model']).agg([min, max]).Fines
```

```python
df.groupby(['Make', 'Model']).std().Fines
```

<h2><center>Агрегация с CarNumber</center></h2>

```python
df.groupby(['CarNumber']).count().Fines.sort_values(ascending=False) # top-3 for check-list
```

```python
df[df.index == df.groupby(['CarNumber']).count().Fines.sort_values(ascending=False).index[0]] 
```

```python
df.groupby(['CarNumber']).sum().Fines.sort_values(ascending=False) # top-1 for check-list
```

```python
df[df.index == df.groupby(['CarNumber']).sum().Fines.sort_values(ascending=False).index[0]]
```

```python
df.groupby('CarNumber').count().Model[df.groupby('CarNumber').count().Model > 1]
```
