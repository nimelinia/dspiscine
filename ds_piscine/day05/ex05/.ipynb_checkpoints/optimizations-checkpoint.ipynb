```python
import pandas as pd
```

```python
import gc
```

<h2><center>Чтение из CSV</center></h2>

```python
fines = pd.read_csv('../data/fines.csv')
```

```python
fines
```

<h2><center>LOOP</center></h2>

```python
def loop(df):
    res = []
    for i in range(len(df)):
        res.append(df.iloc[i].Fines/df.iloc[i].Refund * df.iloc[i].Year)
    return res
```

```python
%%timeit
fines['New_column'] = pd.Series(loop(fines))
```

```python
fines
```

```python
fines.drop('New_column', axis=1)
```

<h2><center>ITERROWS</center></h2>

```python
def iter_rows(df):
    res = []
    for row in df.iterrows():
        res.append(row[1].Fines / row[1].Refund * row[1].Year)
    return res
```

```python
%%timeit
fines['New_column'] = pd.Series(iter_rows(fines))
```

```python
fines
```

```python
fines.drop('New_column', axis=1)
```

<h2><center>LAMBDA FUNC AND APPLY</center></h2>

```python
%%timeit
fines['New_column'] = fines.apply(lambda x: x.Fines / x.Refund * x.Year, axis=1)
```

```python
fines
```

```python
fines.drop('New_column', axis=1)
```

<h2><center>С ПОМОЩЬЮ SERIES</center></h2>

```python
%%timeit
fines['New_column'] = fines.Fines / fines.Refund * fines.Year
```

```python
fines
```

```python
fines.drop('New_column', axis=1)
```

<h2><center>С ПОМОЩЬЮ SERIES И VALUES</center></h2>

```python
%%timeit
fines['New_column'] = fines.Fines.values / fines.Refund.values * fines.Year.values
```

```python
fines
```

```python
fines.drop('New_column', axis=1)
```

<h2><center>DOWNCASTING</center></h2>

```python
fines.info(memory_usage='deep')
```

```python
fines_optimized = fines.copy()
```

```python
fines_optimized
```

```python
fcols = fines_optimized.select_dtypes('float').columns
fines_optimized[fcols] = fines_optimized[fcols].apply(pd.to_numeric, downcast='float')
```

```python
fines_optimized.info(memory_usage='deep')
```

```python
icols = fines_optimized.select_dtypes('integer').columns
fines_optimized[icols] = fines_optimized[icols].apply(pd.to_numeric, downcast='integer')
```

```python
fines_optimized.info(memory_usage='deep')
```

<h2><center>CATEGORIES</center></h2>

```python
objcols = fines_optimized.select_dtypes('object').columns
fines_optimized[objcols] = fines_optimized[objcols].astype('category')
```

```python
fines_optimized.info(memory_usage='deep')
```
