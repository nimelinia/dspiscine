<h2><center>Создание dataframe</center></h2>

```python
df = pd.read_json('../data/auto.json', orient='records')
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df
```

<h2><center>Генерация данных и добавление их в dataframe</center></h2>

```python
df_new = df.sample(n=200, random_state=21, ignore_index=True)
```

```python
df_new['Refund'] = df.sample(n=200, weights='Refund', random_state=13, ignore_index=True).Refund
```

```python
df_new.head()
```

```python
df_new['Fines'] = df.sample(n=200, weights='Fines', random_state=31, ignore_index=True).Fines
```

```python
df_new.head()
```

```python
concat_rows = pd.concat([df, df_new], ignore_index=True)
```

```python
concat_rows
```

```python
concat_rows.count() # для чек-листа
```

<h2><center>Добавление колонки Fine с годом в dataframe</center></h2>

```python
np.random.seed(21)
```

```python
Year = pd.Series(np.random.randint(1980, 2020) for _ in range(925))
```

```python
Year
```

```python
fines = pd.concat([concat_rows, pd.DataFrame(Year, columns=['Year'])], ignore_index=True, axis=1).rename(columns={0: 'CarNumber', 1: 'Refund', 2: 'Fines', 3: 'Make', 4: 'Model', 5: 'Year'})
```

```python
fines
```

```python
fines.count() # для чек-листа
```

<h2><center>Создание dataframe с именами водителей</center></h2>

```python
df_buf = pd.read_json('../data/surname.json')
```

```python
df_buf.rename(columns={0: "NAME", 1: "COUNT", 2: "RANK"}, inplace = True)
```

```python
df_buf = df_buf.drop(index=0)
```

```python
df_buf.tail()
```

```python
Surnames = pd.Series(df_buf.sample(n=len(pd.unique(concat_rows['CarNumber'])), random_state=21, replace=True, ignore_index=True).NAME) 
```

```python
Surnames
```

```python
owners=pd.concat([pd.DataFrame(pd.unique(concat_rows['CarNumber'])), pd.DataFrame(Surnames)], ignore_index=True, axis=1).rename(columns={0: "CarNumber", 1: "SURNAME"})
```

```python
owners
```

```python
len(owners) # для чек-листа
```

```python
fines_new = fines.sample(n=5, random_state=21, ignore_index=True)
```

```python
fines_new['CarNumber'] = pd.Series(['111' + fines_new.CarNumber[x][3:] for x in range(5)])
```

```python
fines = pd.concat([fines, fines_new], ignore_index=True)
```

```python
fines
```

```python
len(fines) # для чек-листа
```

```python
owners.tail(35)
```

```python
owners.drop(index=owners.tail(20).index, inplace=True)
```

```python
owners.tail(20)
```

```python
owners_new = owners.sample(n=3, random_state=21, ignore_index=True)
```

```python
owners_new['CarNumber'] = pd.Series(['777' + owners_new.CarNumber[x][3:] for x in range(3)])
```

```python
# owners = owners.append(owners_new, ignore_index=True)
```

```python
owners = pd.concat([owners, owners_new], ignore_index=True)
```

```python
owners
```

```python
fines.merge(owners, on='CarNumber', how='inner')
```

```python
owners.merge(fines, on='CarNumber', how='outer')
```

```python
owners.merge(fines, on='CarNumber', how='right')
```

```python
owners.merge(fines, on='CarNumber', how='left')
```
