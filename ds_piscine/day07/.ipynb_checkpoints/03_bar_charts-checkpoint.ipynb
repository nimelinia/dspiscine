Что делать, если среднее количество коммитов отличается в рабочий или выходной день?

Сделайте то, что вам нужно сделать, чтобы создать такой график:

- анализируйте только пользователей, а не админов
- размер шрифта и размер инжира остаются прежними
- для каждого часа рассчитайте среднее количество коммитов в рабочие и выходные дни (если не было никаких коммитов в течение часа, не используйте его для расчета среднего), используйте эти значения для своего графика, например: Пн, 17- 18: 5 коммитов, Вт, 17-18: 6 коммитов, ср 17-18: 7 коммитов
- выберите палитру, которая вам действительно нравится, вам не нужно копировать ее из графика выше
- в конце записной книжки Jupyter создайте ячейку уценки и вставьте вопрос
     - «Различается ли динамика по рабочим и выходным дням?», Потому что ответ включает час, когда количество коммитов наибольшее в рабочие дни, и час, когда оно наибольшее в выходные.

```python
con = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
commits = pd.io.sql.read_sql("""
    SELECT *
    FROM (
        SELECT
            date(ch.timestamp) as date,
            cast(strftime('%H',ch.timestamp) as int) as hour,
            COUNT() as commits
        FROM
            checker ch
    where
        uid like 'user_%'
    GROUP BY
        date, hour)
""", con, parse_dates=['timestamp'])

commits.head(5)
```

```python
commits.shape
```

```python
commits['working_day'] = (
    commits.date.apply(lambda date: 'working_day' if pd.to_datetime(date).dayofweek < 5 else 'weekend')
)

commits.head(5)
```

```python
working_day = commits.query("working_day =='working_day'")
weekend =  commits.query("working_day == 'weekend'")

weekend.head(5)
```

```python
working_day = working_day.drop(['working_day','date'], axis=1).sort_values('hour')
working_day.rename(columns={'commits':'working_day'}, inplace=True)
working_day=working_day.groupby('hour').mean().reset_index()

working_day.head(5)
```

```python
weekend = weekend.drop(['working_day','date'], axis=1).sort_values('hour')
weekend.rename(columns={'commits':'weekend'},inplace=True)
weekend=weekend.groupby('hour').mean().reset_index()
weekend.head(5)
```

```python
hours = list(range(24))
hours = pd.DataFrame(hours, columns=['hour'])
```

```python
plot_data = hours.merge(working_day,on='hour', how='left').merge(weekend,on='hour',how='left')

plot_data.head(30)
```

```python
plot_data.info()
```

```python
# https://matplotlib.org/stable/tutorials/colors/colormaps.html

plot = plot_data.plot(
    y=['weekend','working_day'],
    stacked=True,
    figsize=(15, 8),
    title="Commits per hour",
    fontsize=8,
    kind='bar',
    colormap='plasma',
    subplots=True,
    x='hour',
    rot=0
)
```
