```python
import pandas as pd
```

```python
import sqlite3
```

<h2><center>Соединение с БД</center></h2>

```python
connection = sqlite3.connect('../data/checking-logs.sqlite')
```

<h2><center>Запросы к БД</center></h2>

```python
commits = pd.io.sql.read_sql(
    """
    select * from (
        select
            date(checker.timestamp) as date,
            cast(strftime('%H',checker.timestamp) as int) as hour,
            count() as commits
        from
            checker
    where
        uid like '%user_%'
    group by date, hour)
    """, connection, parse_dates='timestamp')
```

```python
commits['type_of_day'] = (
    commits.date.apply(lambda date: 'working_day' if pd.to_datetime(date).dayofweek < 5 else 'weekend')
)
```

```python
commits
```

```python
workdays = commits.query("type_of_day =='working_day'")
weekend = commits.query("type_of_day =='weekend'")
```

```python
workdays = workdays.drop(['type_of_day'], axis=1).groupby('hour').mean().rename(columns={'commits': 'working_day'})
workdays
```

```python
weekend = weekend.drop(['type_of_day'], axis=1).groupby('hour').mean().rename(columns={'commits': 'weekend'})
weekend
```

```python
hours = pd.DataFrame(list(range(24)), columns=['hour'])
```

```python
merge_data = hours.merge(workdays, on='hour', how='left').merge(weekend,on='hour',how='left')
merge_data
```

```python
axes = merge_data.plot(y = ['weekend', 'working_day'], fontsize=8, figsize=(15, 8), title='Commits per hour', xlabel='hour', \
              kind='bar', stacked=True, colormap='Pastel2_r', subplots=True, rot=0)
# axes.patch.set_facecolor('#DCDCDC')
# axes.grid(axis='y')
```
